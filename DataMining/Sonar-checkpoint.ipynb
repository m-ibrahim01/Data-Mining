The following block extracts the predictors and labels from the training and the test set.

Each line extracts variables using the 'iloc' method to access the data by its integer position.

The '.values' attribute will then return the selected data as a numpy array, ready for use.

```python
# Extracting the predictors and labels from the training set and the test 
X_train = train_df.iloc[:, :-1].values
y_train = train_df.iloc[:, -1].values
X_test = test_df.iloc[:, :-1].values
y_test = test_df.iloc[:, -1].values
```

The following block of code defines the function 'minkowski_distance'

This is created to calculate the Minkowski distance between the two points X and Y, using the parameter q, which controls the distance.

```python
def minkowski_distance(x, y, q):                 # Define the Minkowski distance, take three arguments: x, y, q
    return np.sum(np.abs(x - y) ** q) ** (1 / q) # Calculate the Minkowski distance between x and y, then raise it to the power q, sum the results, and take 1/q power of the result for the final distance
```

The following block of code defines the function 'classify' to perform a nearest neigbour classification on the test record, based on the training set, using the Minkowski distance.

```python
def classify(test_record, X_train, y_train, q):                          # Define function 'classify' to take 4 arguments, test_record, X_train, y_train, and q.
    distances = pairwise_distances(X_train, test_record.reshape(1, -1),
                                   metric=lambda x, y:
                                       minkowski_distance(x, y, q))      # Calculate the pairwise distances between the test record an all records in the training set.
    nearest_neighbour_idx = np.argmin(distances)                         # Find the index of the nearest neighbour in the training set
    return y_train[nearest_neighbour_idx]                                # Return the class label of the nearest neighbour in training set
```

The following block of code classifies all the test records using the training data, then stores the predicted labels in the list named 'predictions'.

```python
predictions = []                                              # Initialise the empty list, storing predicted labels
for i in range(len(X_test)):                                  # Set a loop to iterate over test records
    test_record = X_test[i]                                   # select the test record from the X_test array and assign it to test_record
    prediction = classify(test_record, X_train, y_train, q=1) # Call the classify function to classify the test record using X_train, y_train and Minkowski distance, the result is named prediction
    predictions.append(prediction)                            # Add the predicted label to the predictions list
```

The following block of code calculates evaluation metrics on the test set;
- True positives    |   Instances correctly classified as positive by the model.
- False positives   |   Instances incorrectly classified as positive by the model.
- True negatives    |   Instances correctly classified as positive by the model.
- False negatives   |   Instances incorrectly classified as negative by the model.
- Accuracy          |   The proportion of correct predictions amongst all predictions.
- Precision         |   The proportion of true positive predictions among all positive predictions.
- Recall            |   The proportion of true positive predictions among all actual positive instances.
- F1 Score          |   The harmonic mean of the precision and recall, this gives the balanced measure for both metrics and is the most important one.

```python
# Calculate evaluation metrics for the test set
true_positive = np.sum((y_test == 'M') & (np.array(predictions) == 'M'))    # Counts the number of cases where the true label is M, and the predicted label is also M
false_positive = np.sum((y_test == 'R') & (np.array(predictions) == 'M'))   # Counts the number of cases where the true label is R, and the predicted label is M
true_negative = np.sum((y_test == 'R') & (np.array(predictions) == 'R'))    # Counts the number of cases where the true label is R, and the predicted label is also R
false_negative = np.sum((y_test == 'M') & (np.array(predictions) == 'R'))   # Counts the number of cases where the true label is M, and the predicted label is R

accuracy = (true_positive + true_negative) / len(y_test)                    # Calculate the overall accuracy of the classifier
precision = true_positive / (true_positive + false_positive)                # Calculate the precision of the classifier
recall = true_positive / (true_positive + false_negative)                   # Calculate the recall of the classifier
f1_score = 2 * precision * recall / (precision + recall)                    # Calculate F1 score of the classifier, which is the weighted average of precision and recall
```

The following block of code prints out the evaluation metrics on the test set.

```python
# Prints out the evaluation metrics
print("Evaluation metrics on the test set:")
print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1 score:", f1_score)
```

The accuracy of the model is 0.88, this means that it correctly classified for 88.4% of the instances.

The precision is 0.85, which means that out of all the instances predicted as positive, 85.4% were positive.

The recall is 0.95, which means that out of all the actual positive instances, the model correctly identified 94.6%.

The F1 score is 0.90, which is the mean of the precision and recall, showing the balanced measure of the two metrics.


The evaluation metrics on the test set show that the model is performing very well.
The high recall value indicates to us that the model correctly identifies most of the actual positive instances, and the high precision indicates that it isnt producing many false positives.


<h3>The following block of code performs grid search to find the best value of the hyperparameter q.

(*q = the exponent in the Minkowski distance formula*)<h3>



```python
results = []                                                                      # Create an empty list called results
for q in range(1, 21):                                                            # Start a loop to iterate over q, values from 1 to 20
    predictions = []                                                              # Create an empty list called predictions
    
    for i in range(len(y_test)):                                                  # Start a loop to iterate over i, values from 0 to the length of y_test
        test_record = X_test[i]                                                   # Setting test_record to the 'i' element of the X_test
        prediction = classify(test_record, X_train, y_train, q=q)                 # Uses 'classify' to make a prediction for the test_record, using the value 'q' specified in the outer loop
        predictions.append(prediction)                                            # Add the predicted label to the predictions list
        
    true_positive = np.sum((y_test == 'M') & (np.array(predictions) == 'M'))      # Counts the number of true positives by comparing y_test to predictions, counts cases where both are = 'M'
    false_positive = np.sum((y_test == 'R') & (np.array(predictions) == 'M'))     # Counts the number of false positives by comparing y_test to predictions, counts cases where true label = 'R', predicted = 'M'
    true_negative = np.sum((y_test == 'R') & (np.array(predictions) == 'R'))      # Counts the number of true negatives by comparing y_test to predictions, counts cases where both are = 'R'
    false_negative = np.sum((y_test == 'M') & (np.array(predictions) == 'R'))     # Counts the number of false negatives by comparing y_test to predictions, counts cases where true label = 'M', predicted = 'R'
    
    accuracy = (true_positive + true_negative) / len(y_test)                      # Calculate the accuracy of the predictions
    precision = true_positive / (true_positive + false_positive)                  # Calculate the precision of the predictions
    recall = true_positive / (true_positive + false_negative)                     # Calculate the recall of the predictions
    f1 = 2 * precision * recall / (precision + recall)                            # Calculate the F1 score of the predictions
    results.append([q, accuracy, precision, recall, f1])                          # Appends a list of the 'q' value and the four evaluation metrics
    
headers = ['q', 'Accuracy', 'Precision', 'Recall', 'F1']                          # Creates a list of column headers for the table to be printed
print(tabulate(results, headers=headers, floatfmt='.4f'))                         # Prints the table of the evaluation metrics for each value of 'q', displayed using tabulate function from tabulate library
```

From the table, we can see that the best performance was achieved by q=2, with these results;
- Accuracy = 0.8986
- F1 Score = 0.9114

As q increased beyond 2 the accuracy and the F1 score decreased, this means that the model became less effective at classifying the instances.

Overfitting is most likely the reason why this occured, however another possible reason is the Curse of Dimensionality.

This refers to the problem of having a high number of features in a dataset which makes it difficult to analyse and model the data properly. As the number of features increased, the amount of data required to represent the distribution of the data grew exponentially, this is most likely why we had an issue of overfitting here. It affected the performance of the K nearest neighbours algorithm because as the number of features increased, the distance between the points in the feature becomes less informative.

```python
# Convert the results to a pandas DataFrame
df_results = pd.DataFrame(results, columns=headers)
plt.figure(figsize=(10, 6))

# Create the line plot
sns.lineplot(data=df_results, x='q', y='F1')
plt.xlabel('Q')
plt.ylabel('F1')
plt.show()
```

The graph demonstrates that the relationship between the variables is consistent.
